```python
%sql
CREATE CATALOG IF NOT EXISTS icebase
  COMMENT 'Idaho Mashers Hockey Analytics Platform';

USE CATALOG icebase;

CREATE SCHEMA IF NOT EXISTS bronze
  COMMENT 'Raw ingestion layer — unmodified source data';

CREATE SCHEMA IF NOT EXISTS silver
  COMMENT 'Cleaned and conformed dimensional tables';

CREATE SCHEMA IF NOT EXISTS gold
  COMMENT 'Aggregated business-ready tables and ML outputs';

-- Verify creation
SHOW SCHEMAS IN icebase;
```

```python
%sql
USE CATALOG icebase;
USE SCHEMA bronze;

-- Volumes = managed file storage in Unity Catalog
-- These paths are where the simulator will drop JSON/CSV files

CREATE VOLUME IF NOT EXISTS landing_tickets
  COMMENT 'Landing zone for simulated ticket transaction files';

CREATE VOLUME IF NOT EXISTS landing_customers
  COMMENT 'Landing zone for simulated new customer signup files';

-- Verify volumes
SHOW VOLUMES IN icebase.bronze;
```
